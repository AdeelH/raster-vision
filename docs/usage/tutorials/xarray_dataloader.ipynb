```python
import numpy as np
from xarray import DataArray
import torch
from torch.utils.data import DataLoader
```

```python
class DataArrayDataset:
    def __init__(self, arr):
        self.arr = arr

    def __getitem__(self, i):
        return self.arr[i].to_numpy()

    def __len__(self):
        return self.arr.shape[0]
```

```python
import multiprocessing

# Define a function that each process will execute
def worker_function(process_num):
    print(f"Worker {process_num} is doing some work")

if __name__ == "__main__":
    # Create a list of processes
    processes = []

    # Define the number of processes you want to create
    num_processes = 4

    # Start and run the processes
    for i in range(num_processes):
        process = multiprocessing.Process(target=worker_function, args=(i,))
        processes.append(process)
        process.start()

    # Wait for all processes to complete
    for process in processes:
        process.join()

    print("All processes have completed")
```
